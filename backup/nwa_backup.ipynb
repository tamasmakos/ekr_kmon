```python
# Keeping only the year from the published_at column
data['published_at'] = pd.to_datetime(data['published_at']).dt.year
```

```python
# Columns for grouping
grouping_columns = [
    'buyer_type', 'buyer_activity', 'contract_type', 'place', 'winner_kkv', 'number_of_bids', 'eu_fund', 'contract_awarded',
    'quality_criteria', 'cost_criteria', 'price_criteria'
]

# Creating aggregated DataFrame
aggregated_data = data.groupby(grouping_columns).agg(
    Unique_Procurement_IDs=('procurement_id', 'nunique'),
    Sum_of_Prices=('price', 'sum')
).reset_index()

# Calculate the average price
aggregated_data['Average_Price'] = aggregated_data['Sum_of_Prices'] / aggregated_data['Unique_Procurement_IDs']

# Sort and display the results
sorted_data = aggregated_data.sort_values(by='Unique_Procurement_IDs', ascending=False).head(1000)

sorted_data.head(10)
```

```python
import pandas as pd
import networkx as nx

# Load the main network data and network categories
network_data = data.copy()
network_categories = sorted_data.copy()

# Provided filter_data function
def filter_data(data, buyer_type, buyer_activity, contract_type, place, winner_kkv, published_at, number_of_bids, eu_fund, contract_awarded, quality_criteria, cost_criteria, price_criteria):
    return data[
        (data['buyer_type'] == buyer_type) &
        (data['buyer_activity'] == buyer_activity) &
        (data['contract_type'] == contract_type) &
        (data['place'] == place) &
        (data['winner_kkv'] == winner_kkv) &
        (data['published_at'] == published_at) &
        (data['number_of_bids'] == number_of_bids) &
        (data['eu_fund'] == eu_fund) &
        (data['contract_awarded'] == contract_awarded) &
        (data['quality_criteria'] == quality_criteria) &
        (data['cost_criteria'] == cost_criteria) &
        (data['price_criteria'] == price_criteria)
    ]

# Function to create a bipartite network from filtered data
# Winners = 0, Buyers = 1
# They are connected if they have a common procurement_id
def create_network(data):
    network = nx.Graph()
    for i, row in data.iterrows():
        network.add_node(row['winner'], bipartite=0)
        network.add_node(row['buyer'], bipartite=1)
        network.add_edge(row['winner'], row['buyer'], procurement_id=row['procurement_id'])
    return network

# Divide the network data into subnetworks based on categories and store them
subnetworks = {}

for i, row in network_categories.iterrows():
    filtered_data = filter_data(
        network_data, row['buyer_type'], row['buyer_activity'], row['contract_type'], row['place'], 
        row['winner_kkv'], row['published_at'], row['number_of_bids'], row['eu_fund'], 
        row['contract_awarded'], row['quality_criteria'], row['cost_criteria'], row['price_criteria']
    )

    subnetwork_name = f"network_{i}"
    subnetworks[subnetwork_name] = create_network(filtered_data)

# Check for node overlaps between subnetworks
overlap_info = {}

for net1_name, net1 in subnetworks.items():
    for net2_name, net2 in subnetworks.items():
        if net1_name != net2_name:
            common_nodes = set(net1.nodes()).intersection(set(net2.nodes()))
            overlap_info[(net1_name, net2_name)] = len(common_nodes)
```
