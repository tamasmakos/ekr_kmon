```python
import networkx as nx

# Load the main network data and network categories
network_data = data.copy()
network_categories = sorted_data.copy()

# Provided filter_data function
def filter_data(data, buyer_type, buyer_activity, contract_type, winner_kkv, eu_fund):
    return data[
        (data['buyer_type'] == buyer_type) &
        (data['buyer_activity'] == buyer_activity) &
        (data['contract_type'] == contract_type) &
        (data['winner_kkv'] == winner_kkv) &
        (data['eu_fund'] == eu_fund)
    ]

# Function to create a bipartite network from filtered data
# Winners = 0, Buyers = 1
# They are connected if they have a common procurement_id
def create_network(data):
    network = nx.Graph()
    for i, row in data.iterrows():
        network.add_node(row['winner'], bipartite=0)
        network.add_node(row['buyer'], bipartite=1)
        network.add_edge(row['winner'], row['buyer'], procurement_id=row['contract_id'])

    return network

# Divide the network data into subnetworks based on categories and store them
subnetworks = {}

for i, row in network_categories.iterrows():
    filtered_data = filter_data(
        network_data, row['buyer_type'], row['buyer_activity'], row['contract_type'],
        row['winner_kkv'], row['eu_fund']
    )

    subnetwork_name = f"network_{i}"
    subnetworks[subnetwork_name] = {
        'filtered_data': filtered_data,
        'filter_details': {
            'buyer_type': row['buyer_type'],
            'buyer_activity': row['buyer_activity'],
            'contract_type': row['contract_type'],
            'winner_kkv': row['winner_kkv'],
            'eu_fund': row['eu_fund']
        },
        'network': create_network(filtered_data)
    }
```

```python
import plotly.graph_objects as go
import networkx as nx
import matplotlib.cm as cm
import numpy as np
from matplotlib import colors as mcolors

def create_3d_graph_plotly(df, G, buyer_communities_list, winner_communities_list, centrality, filter_details):
    node_x, node_y, node_z, node_color, node_text = [], [], [], [], []
    edge_x, edge_y, edge_z = [], [], []

    # Community colors
    total_communities = len(buyer_communities_list) + len(winner_communities_list)
    community_colors = cm.rainbow(np.linspace(0, 1, total_communities))

    # Assign colors to nodes
    node_color_map = {}
    for idx, comm in enumerate(buyer_communities_list + winner_communities_list):
        color = community_colors[idx]
        for node in comm:
            node_color_map[node] = color

    # Unified 3D layout
    pos = nx.spring_layout(G, dim=3, seed=42)

    # Assign positions and colors
    for node in G.nodes():
        x, y, z = pos[node]
        node_x.append(x)
        node_y.append(y)
        node_z.append(z)
        color = node_color_map.get(node, 'grey')
        node_color.append(mcolors.to_hex(color))
        node_text.append(f'{node}')

    # Normalize centrality values for node sizes
    # Flatten the centrality measures into a single list
    flat_centrality = [value for community in centrality.values() for value in community.values()]

    # Find the maximum centrality value
    max_centrality = max(flat_centrality)
    min_centrality = min(centrality.values())
    size_range = [10, 50]  # Adjust size range as needed

    # Determine node sizes based on community-level centrality
    node_sizes = []
    for node in G.nodes():
        for community_id, community_nodes in enumerate(communities_list):
            if node in community_nodes:
                centrality_value = community_centrality[community_id].get(node, 0)
                break
        size = (centrality_value - min_centrality) / (max_centrality - min_centrality) * (size_range[1] - size_range[0]) + size_range[0]
        node_sizes.append(size)
        
    # Create node trace with updated sizes
    node_trace = go.Scatter3d(x=node_x, y=node_y, z=node_z, mode='markers', 
                            marker=dict(color=node_color, size=node_sizes, colorscale='Rainbow'),
                            text=node_text, hoverinfo='text')


    for edge in G.edges():
        x0, y0, z0 = pos[edge[0]]
        x1, y1, z1 = pos[edge[1]]
        edge_x.extend([x0, x1, None])
        edge_y.extend([y0, y1, None])
        edge_z.extend([z0, z1, None])

    edge_trace = go.Scatter3d(x=edge_x, y=edge_y, z=edge_z, mode='lines', 
                              line=dict(color='black', width=1), hoverinfo='none')

    # Create the 3D plot
    fig = go.Figure(data=[edge_trace, node_trace])

    # Format the filter details as a string for annotation
    filter_details_text = "<br>".join([f"{k}: {v}" for k, v in filter_details.items()])

    # Add an annotation for filter details
    fig.add_annotation(
        text=filter_details_text,  # Formatted filter details
        align='left',
        showarrow=False,
        xref='paper',
        yref='paper',
        x=0,  # Position on the left
        y=1,  # Position at the top
        bordercolor='black',
        borderwidth=2,
        bgcolor='white',
        font=dict(
            size=10
        )
    )

    # Update layout with title and legend
    fig.update_layout(
        title='K-MONITOR - EKR SUBNETWORK',
        legend_title_text='Node Type',
        showlegend=False,
        legend=dict(
            itemsizing='constant',
            traceorder='normal',
            font=dict(
                family='sans-serif',
                size=12,
                color='black'
            ),
            bgcolor='LightSteelBlue',
            bordercolor='Black',
            borderwidth=2
        ),
        scene=dict(
            camera=dict(
                up=dict(x=0, y=1, z=0),  # Align 'up' direction with y-axis
                center=dict(x=0, y=0, z=0),
                eye=dict(x=1.25, y=1.25, z=1.25)  # Camera position adjustments
            )
        )
    )

    return fig
```

```python
import os
import community as community_louvain
import networkx as nx
import plotly.offline as py_offline


# Function to detect communities
def detect_communities(G):
    partition = community_louvain.best_partition(G)
    communities = {}
    for node, community_id in partition.items():
        communities.setdefault(community_id, []).append(node)
    return list(communities.values())

def create_visualization_and_save(df, G, buyer_communities, winner_communities, filter_details, centrality, filename):
    fig = create_3d_graph_plotly(df, G, buyer_communities, winner_communities, centrality, filter_details)
    py_offline.plot(fig, filename=filename, auto_open=False)

def calculate_community_centrality(G, communities):
    community_centrality = {}
    for community_id, community_nodes in enumerate(communities):
        subgraph = G.subgraph(community_nodes)
        centrality = nx.degree_centrality(subgraph)  # or other centrality measures
        community_centrality[community_id] = centrality
    return community_centrality

output_directory = 'network_visualizations'

# Process each subnetwork
for network_name, subnetwork in subnetworks.items():
    # Extract the filtered data and network from the subnetwork
    filtered_data = subnetwork['filtered_data']
    network = subnetwork['network']
    
    # Project bipartite graph to create separate graphs for buyers and winners
    buyers_graph = nx.bipartite.projected_graph(network, [n for n, d in network.nodes(data=True) if d.get('bipartite', '') == 1])
    winners_graph = nx.bipartite.projected_graph(network, [n for n, d in network.nodes(data=True) if d.get('bipartite', '') == 0])
    
    # Calculate centrality measures (e.g., degree centrality) on the projected graphs
    buyer_centrality = nx.degree_centrality(buyers_graph)
    winner_centrality = nx.degree_centrality(winners_graph)

    # Detect communities and calculate centrality measures for each community
    buyer_communities = detect_communities(buyers_graph)
    winner_communities = detect_communities(winners_graph)

    subnetwork['buyer_community_centrality'] = calculate_community_centrality(buyers_graph, buyer_communities)
    subnetwork['winner_community_centrality'] = calculate_community_centrality(winners_graph, winner_communities)

    # Extract filter details for this subnetwork
    filter_details = subnetwork['filter_details']

    # Decide which community-level centrality measure to use for visualization
    centrality_to_use = subnetwork['winner_community_centrality']  # or 'winner_community_centrality'

    # Extract communities for the chosen type (buyers or winners)
    communities_to_use = winner_communities  # or winner_communities

    # Create and save the visualization with community-level centrality-based node sizes
    filename = os.path.join(output_directory, f"{network_name}_visualization.html")
    create_visualization_and_save(network_data, network, buyer_communities, winner_communities, filter_details, centrality_to_use, filename)
```
